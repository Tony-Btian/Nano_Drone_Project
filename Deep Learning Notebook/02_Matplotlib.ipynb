## Matplotlib
在深度学习的实验中，图形的绘制和数据的可视化非常重要。Matplotlib 是用于绘制图形的库，使用 Matplotlib 可以轻松绘制图形和实现数据的可视化。

### 1. 绘制简单图形
绘制 sin。这里使用 NumPy 的 arange 方法生成了 [0, 0.1, 0.2, ..., 5.8, 5.9] 的数据，将其设为 x 。对 x 的各个元素，应用 NumPy 的 sin 函数 np.sin()，将 x、y 的数据传给 plt.plot 方法，然后绘制图形。


```python
import numpy as np  # type: ignore
import matplotlib.pyplot as plt # type: ignore

# 生成数据
x = np.arange(0, 6, 0.1) # 以0.1为单位，生成0到6的数据
y = np.sin(x)

# 绘制图形
plt.plot(x, y)
plt.show()
```

#### 1.1 追加 cos 函数的图形
在原有图形的基础上追加 cos 函数的图形，尝试使用 pyplot 的添加标题和 x 轴标签名等其他功能。

```python
import numpy as np  # type: ignore
import matplotlib.pyplot as plt # type: ignore

# 生成数据
x = np.arange(0, 6, 0.1) # 以0.1为单位，生成0到6的数据
y1 = np.sin(x)
y2 = np.cos(x)

# 绘制图形
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos") # 用虚线绘制
plt.xlabel("x") # x轴标签
plt.ylabel("y") # y轴标签
plt.title('sin & cos')
plt.legend()
plt.show()
```

#### 1.2 显示图像
pyplot 中提供了用于显示图像的方法 imshow()。另外，可以使用 matplotlib.image 模块的 imread() 方法读入图像。

```python
import matplotlib.pyplot as plt # type: ignore
from matplotlib.image import imread # type: ignore

img = imread('Material/Lusi.jpg') # 读入图像（设定合适的路径）
plt.imshow(img)

plt.show()
```
